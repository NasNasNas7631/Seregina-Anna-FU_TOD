Считайте, что в слове есть опечатка, если данное слово не содержится в списке `words`. 

```python
import nltk
from nltk.metrics.distance import jaccard_distance
from nltk.util import ngrams
import nltk
from nltk.corpus import words
from nltk.metrics.distance import edit_distance
from nltk.util import ngrams
nltk.download('words')
import pandas
nltk.download('punkt')
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
words = []
with open("litw-win.txt", "r", encoding = "windows-1251") as f:
  for l in f:
    words.append(l.split()[1])
```

```python
text = '''с велечайшим усилием выбравшись из потока убегающих людей Кутузов со свитой уменьшевшейся вдвое поехал на звуки выстрелов русских орудий'''
```

```python
for i in word_tokenize(text, language="russian"):
    if i not in words:
      dis = [edit_distance(i, j) for j in words]
      print(words[dis.index(min(dis))], end = ' ')
    else:
      print(i, end = " ")
  
```

2. Разбейте текст из формулировки задания 1 на слова; проведите стемминг и лемматизацию слов.

```python
text = "Считайте слова из файла litw-win.txt и запишите их в список words. В заданном предложении исправьте все опечатки, заменив слова с опечатками на ближайшие (в смысле расстояния Левенштейна) к ним слова из списка words. Считайте, что в слове есть опечатка, если данное слово не содержится в списке words."
```

```python
snowball = SnowballStemmer(language="russian")
texti = word_tokenize(text, language="russian")
morph = pymorphy2.MorphAnalyzer()
stem = []
lim = []
for i in texti:
  stem.append(snowball.stem(i))
  lim.append(morph.parse(i)[0].normal_form)
print(stem)
print(lim)
```

2.2. Удалите стоп-слова из описаний рецептов. Какую долю об общего количества слов составляли стоп-слова? Сравните топ-10 самых часто употребляемых слов до и после удаления стоп-слов.

```python
 from nltk.corpus import stopwords 
 nltk.download('stopwords') 
```

```python
import pandas as pd
descript = pd.read_csv('preprocessed_descriptions.csv', index_col = 'Unnamed: 0')
list_descript = descript['preprocessed_descriptions'].to_list()
words = [k for j in [word_tokenize(str(i)) for i in list_descript] for k in j]
```

```python
wordss = [i for i in words[:100] if not i in stopwords.words()]
```

```python
import collections
print('Всего слов:', len(words), '\nСлов без стоп-слов', len(wordss))
print('Доля стоп-слов', ((len(words) - len(wordss))/len(words))*0.01)
c1 = collections.Counter(words)
c2 = collections.Counter(wordss)
print('Топ-5 слов до удаления', c1.most_common(5))
print('Топ-5 слов после удаления', c2.most_common(5))
```
