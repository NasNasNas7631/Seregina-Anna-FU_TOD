### pickle

2.1 На основе файла `contributors_sample.json` создайте словарь следующего вида: 
```
{
    должность: [список username людей, занимавших эту должность]
}
```

````python
import json
import numpy as np
with open ("contributors_sample.json", "r") as f:
  abc = json.load(f)
b = set(i for j in [i['jobs'] for i in abc] for i in j)
u = {i['username']:i['jobs'] for i in abc}
c = dict()
print(b)
for i in u:
  for j in u[i]:
    if j in c:
      c[j].append(i)
    else:
      c[j] = [i]
for k,v in c.items():
  print(k, ":", v)
````

````python
import json
from bs4 import BeautifulSoup
with open ("steps_sample.xml", "r") as f:
  abc = BeautifulSoup(f, "xml")
````

````python
a = []
b = []
for per in abc.find_all("recipe"):
  a = [i.next for i in per.steps.find_all("step")]
  b.append({per.find("id").next: a})
with open('steps_sample.json', 'w') as f:
  json.dump(b, f, indent = 2)
````

````python
steps_len = []
dict_nsteps_id = {}
for recipe in abc.recipes.find_all('recipe'):
  num_steps = len([step.next for step in abc.steps.find_all('step')])
  if num_steps not in dict_nsteps_id:
    dict_nsteps_id[num_steps] = [id.next for id in abc.find_all('id')]
  else:
    dict_nsteps_id[num_steps] += [id.next for id in abc.find_all('id')]
for i in dict_nsteps_id:
  print(i, ':', dict_nsteps_id[i])
````
